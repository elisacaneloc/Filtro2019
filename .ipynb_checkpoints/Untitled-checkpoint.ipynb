```python
import medicion as md
import numpy as np
from scipy.stats import kstest
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
def model(x,a,b,c):
    return -a*np.exp(-x*b) + c
```

```python
D, C, Vbias, Speed = md.readmeasure('C:/Users/Elisa Canelo/Documents/GitHub/Filtro2019/molecule/16-20/scan170429_4_{0}.dat',5000)
```

```python
x_data = D[0]
y_data = C[0]
a0 = 1, 1, 1
a_opt, a_covar = curve_fit(model,x_data , y_data, a0)
a_opt
```

```python
x = np.linspace(-2.5, 4.5, 100)
plt.semilogy(x, model(x, *a_opt))
```
